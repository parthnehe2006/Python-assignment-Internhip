```python
from google import genai
from google.genai import types
from google.colab import userdata

# Load API Key from Colab Secrets
API_KEY = userdata.get("Gemini_API_Key")

# Create Gemini Client
client = genai.Client(api_key=API_KEY)

def generate_blog(topic, tone, word_count):

    prompt = f"""
You are an experienced blog writer.

Write a {tone} blog on the topic "{topic}".

Requirements:
- Around {word_count} words
- Catchy Title
- Introduction Hook
- 2-3 Body Sections
- Conclusion
"""

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt,
        config=types.GenerateContentConfig(
            temperature=0.7
        )
    )

    blog = response.text

    # Save blog to file
    with open("blog_output.txt", "w", encoding="utf-8") as file:
        file.write(blog)

    return blog

topic = input("Enter Topic: ")
tone = input("Enter Tone (formal/casual/Gen Z/professional): ")
word_count = input("Enter Word Count: ")

blog = generate_blog(topic, tone, word_count)

print("\n===== BLOG GENERATOR =====\n")
print(f'Topic: "{topic}"')
print(f"Tone: {tone}")
print(f"Word count: {word_count}\n")

print(blog)

print("\nBlog saved as blog_output.txt")
```
