```python
import google.generativeai as genai
from google.colab import userdata

genai.configure(api_key=userdata.get("Gemini_API_Key"))

knowledge_base = """
Business Name: TechWorld Electronics

FAQs

1. What products do you sell?
We sell laptops, smartphones, tablets, smartwatches, headphones, and accessories.

2. What are your working hours?
Monday to Saturday, 9:00 AM to 8:00 PM.

3. Are you open on Sundays?
No.

4. Where are you located?
MG Road, Bengaluru.

5. Do you provide home delivery?
Yes, within Bengaluru.

6. What payment methods are accepted?
Cash, UPI, Debit Card, Credit Card, and Net Banking.

7. Do you offer EMI?
Yes, EMI is available on selected products.

8. What is your return policy?
Products can be returned within 7 days if they are unopened.

9. What is your warranty policy?
Products come with the manufacturer's warranty.

10. How can I contact customer support?
Call 9876543210 or email support@techworld.com.

11. Do you repair laptops?
Yes, we provide laptop repair services.

12. Do you sell gaming laptops?
Yes, we have gaming laptops from ASUS, Acer, MSI, and Lenovo.

13. Do you have student discounts?
Yes, valid student ID is required.

14. Can I preorder products?
Yes, preorders are accepted for selected products.

15. Do you provide installation services?
Yes, for selected electronic devices.
"""
system_instruction = f"""
You are the FAQ chatbot for TechWorld Electronics.

Use ONLY the information provided below.

{knowledge_base}

Rules:
1. Answer only from the knowledge base.
2. If the answer is not available, politely say:
"I'm sorry, I can only answer questions related to TechWorld Electronics."
3. Be short and professional.
"""
model = genai.GenerativeModel(
    model_name="gemini-2.5-flash",
    system_instruction=system_instruction,
    generation_config={
        "temperature":0.2
    }
)

chat = model.start_chat(history=[])
print("========== FAQ CHATBOT ==========")
print("Type 'quit' or 'exit' to stop.\n")

while True:

    user_input = input("You: ")

    if user_input.lower() in ["quit", "exit"]:
        print("Bot: Thank you! Have a great day.")
        break

    response = chat.send_message(user_input)

    print("\nBot:", response.text)
    print()
```
